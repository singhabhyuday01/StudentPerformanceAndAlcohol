# Student Performance and Alcohol

### An analysis of Student’s performance dataset from the UCI ML Repository.

## Introduction

The UCI Machine Learning Repository provides a data set of students’ performance in secondary education of two Portuguese schools in two subjects: Mathematics and Portuguese language. It mentions various attributes such as student grades, demographic, social, and school-related features.

## Business Understanding

One of the reasons for my interest in this particular data set was the alcohol consumption attribute. This data set shows the workday and weekend alcohol consumption by students in a five-level marker- from 1 to 5. Alcoholism is a problem that many people have to fight against their whole life. The fact that secondary school students consume alcohol is a concern worth spending time over. Not only it damages their health at a young age, but it could also lead to addiction and subsequently hamper their chances of leading a happy and successful life.

### Project Goal
Along with exploring the data, I aim to find more information about the following questions:
##### 1. Of those who performed poorly, how many had high alcohol consumption?
##### 2. Do students drink more, less, or the same on weekends as they do on workdays?
##### 3. Does parents’ education affects student’s grades?

## Data Understanding

### Attributes

Attributes for both student-mat.csv (Math course) and student-por.csv (Portuguese language course) datasets:

1 school - student's school (binary: "GP" - Gabriel Pereira or "MS" - Mousinho da Silveira) <br>
2 sex - student's sex (binary: "F" - female or "M" - male) <br>
3 age - student's age (numeric: from 15 to 22) <br>
4 address - student's home address type (binary: "U" - urban or "R" - rural) <br>
5 famsize - family size (binary: "LE3" - less or equal to 3 or "GT3" - greater than 3) <br>
6 Pstatus - parent's cohabitation status (binary: "T" - living together or "A" - apart) <br>
7 Medu - mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education) <br>
8 Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education) <br>
9 Mjob - mother's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other") <br>
10 Fjob - father's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other") <br>
11 reason - reason to choose this school (nominal: close to "home", school "reputation", "course" preference or "other") <br>
12 guardian - student's guardian (nominal: "mother", "father" or "other") <br>
13 traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour) <br>
14 studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours) <br>
15 failures - number of past class failures (numeric: n if 1<=n<3, else 4) <br>
16 schoolsup - extra educational support (binary: yes or no) <br>
17 famsup - family educational support (binary: yes or no) <br>
18 paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no) <br>
19 activities - extra-curricular activities (binary: yes or no) <br>
20 nursery - attended nursery school (binary: yes or no) <br>
21 higher - wants to take higher education (binary: yes or no) <br>
22 internet - Internet access at home (binary: yes or no) <br>
23 romantic - with a romantic relationship (binary: yes or no) <br>
24 famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent) <br>
25 freetime - free time after school (numeric: from 1 - very low to 5 - very high) <br>
26 goout - going out with friends (numeric: from 1 - very low to 5 - very high) <br>
27 Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high) <br>
28 Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high) <br>
29 health - current health status (numeric: from 1 - very bad to 5 - very good) <br>
30 absences - number of school absences (numeric: from 0 to 93)

These grades are related with the course subject, Math or Portuguese: <br>
31 G1 - first period grade (numeric: from 0 to 20) <br>
31 G2 - second period grade (numeric: from 0 to 20) <br>
32 G3 - final grade (numeric: from 0 to 20, output target) <br>

Additional note: there are several (382) students that belong to both datasets. <br>
These students can be identified by searching for identical attributes that characterize each student <br>

## Question 1: Of those who performed poorly, how many had high alcohol consumption?

To answer this question, we first need to define what we mean by “poor performance” and “high alcohol consumption”.<br>
For performance, we use the G3 attribute of the data. Considering the Erasmus grade conversion system (European exchange program that enables student exchange in 31 countries), we assign grades as follows:

<img src="grades.png" />

We define bad performance as grades 2 and 1. The table is as follows with Grade and Performance attributes:

```python
def grades(val, field='G3'):
    marks = val[field]
    if marks >= 16:
        return 5
    elif marks >= 14:
        return 4
    elif marks >= 12:
        return 3
    elif marks >= 10:
        return 2
    else:
        return 1
```

```python
dfMath['Grade'] = dfMath.apply(grades, axis=1)
dfPort['Grade'] = dfPort.apply(grades, axis=1)
```

```python
calc_performance = lambda val: 'Good' if val>=3 else 'Bad'
```

```python
dfMath['Performance'] = dfMath['Grade'].apply(calc_performance)
dfMath.sample(5)
```

```python
dfPort['Performance'] = dfPort['Grade'].apply(calc_performance)
dfPort.sample(5)
```

For alcohol consumption, we define high alcohol consumption to be equal to or greater than 3 on either workday or on weekends.

```python
highAlc = lambda val: val['Dalc']>=3 or val['Walc']>=3
lowAlc = lambda val: val['Dalc']<3 and val['Walc']<3
```

Analyzing the data for the two subjects, we create a data frame to store and describe the values.

```python
alc_perf_summary = pd.DataFrame(columns=['highAlc_Math','lowAlc_Math','highAlc_Port','lowAlc_Port'])
alc_perf_summary
```

```python
highAlcMath = dfMath[dfMath.apply(highAlc, axis=1)]
print("Mathematics Students with high alcohol content: {}".format(highAlcMath.shape[0]))
highAlcMathSummary = highAlcMath.groupby('Performance')['Dalc'].count()*100/highAlcMath.shape[0]
print(highAlcMathSummary)
alc_perf_summary['highAlc_Math'] = highAlcMathSummary.round(decimals=2)
alc_perf_summary
```

```python
lowAlcMath = dfMath[dfMath.apply(lowAlc, axis=1)]
print("Mathematics Students with low alcohol content: {}".format(lowAlcMath.shape[0]))
lowAlcMathSummary = lowAlcMath.groupby('Performance')['Dalc'].count()*100/lowAlcMath.shape[0]
print(lowAlcMathSummary)
alc_perf_summary['lowAlc_Math'] = lowAlcMathSummary.round(decimals=2)
alc_perf_summary
```

Visualizing the results for Mathematics using a pie-chart:

```python
fig, axs = plt.subplots(1, 2, figsize=(14,14))
axs[0].pie(highAlcMath.groupby('Grade')['Dalc'].count(), 
            labels=['fail','sufficient','satisfactory','good','excellent'],
        autopct='%1.1f%%', explode=(0.05,0.05,0,0,0),
          colors=['salmon', 'orange', 'lightskyblue', 'yellowgreen', 'lightgreen'])
axs[0].set_title("High Alcohol Consumption")
axs[1].pie(lowAlcMath.groupby('Grade')['Dalc'].count(), 
           labels=['fail','sufficient','satisfactory','good','excellent'],
        autopct='%1.1f%%', explode=(0.05,0.05,0,0,0), startangle=30,
          colors=['salmon', 'orange', 'lightskyblue', 'lightgreen', 'yellowgreen'])
axs[1].set_title("Low Alcohol Consumption");
```

Doing the same process for Portuguese students:

```python
highAlcPort = dfPort[dfPort.apply(highAlc, axis=1)]
print("Portuguese Students with high alcohol content: {}".format(highAlcPort.shape[0]))
highAlcPortSummary = highAlcPort.groupby('Performance')['Dalc'].count()*100/highAlcPort.shape[0]
print(highAlcPortSummary)
alc_perf_summary['highAlc_Port'] = highAlcPortSummary.round(decimals=2)
alc_perf_summary
```

```python
lowAlcPort = dfPort[dfPort.apply(lowAlc, axis=1)]
print("Portuguese Students with low alcohol content: {}".format(lowAlcPort.shape[0]))
lowAlcPortSummary = lowAlcPort.groupby('Performance')['Dalc'].count()*100/lowAlcPort.shape[0]
print(lowAlcPortSummary)
alc_perf_summary['lowAlc_Port'] = lowAlcPortSummary.round(decimals=2)
alc_perf_summary
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14,14))
axs[0].pie(highAlcPort.groupby('Grade')['Dalc'].count(), 
            labels=['fail','sufficient','satisfactory','good','excellent'],
        autopct='%1.1f%%', explode=(0.07,0.07,0,0,0),
          colors=['salmon', 'orange', 'lightskyblue', 'yellowgreen', 'lightgreen'])
axs[0].set_title("High Alcohol Consumption")
axs[1].pie(lowAlcPort.groupby('Grade')['Dalc'].count(), labels=['fail','sufficient','satisfactory','good','excellent'],
        autopct='%1.1f%%', explode=(0.07,0.07,0,0,0), startangle=30,
          colors=['salmon', 'orange', 'lightskyblue', 'yellowgreen', 'lightgreen'])
axs[1].set_title("Low Alcohol Consumption");
```

### Result

I exploded the values for fail and sufficient grades in order to give a better idea.

In both subjects, it is evident from the above pie charts that the majority of those who performed poorly had high alcohol consumption and the proportion of low-performing students goes down with lower alcohol consumption. This proves that there is a correlation between high alcohol consumption and low grades.

## Question 2: Do students drink more, less, or the same on weekends as they do on workdays?

Analyzing the data, we can know how many students drink more or less than they do on the weekdays or just as same.

For this question, we would merge the records from two tables on common attributes: <br>
[“school”, “sex”,” age”, “address”, “famsize”, “Pstatus”, “Medu”, “Fedu”, “Mjob”, “Fjob”, “reason”, “nursery”, “internet”].

We perform the set union operation on the two tables; the equality condition being the data in the common columns. The pandas library does not provide a set union operator. However, we can achieve the similar result by applying the concatenation operation and then dropping the duplicates based on the common attributes above.<br>
We know there are 382 students which are common in both.

```python
common_cols = ["school","sex","age","address","famsize","Pstatus",
               "Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]
students_common = 382
```

```python
print("Math: {}, Port: {}, Both: {}".format(dfMath.shape[0], dfPort.shape[0], students_common))
print("{} + {} - {} = {}".format(dfMath.shape[0], dfPort.shape[0], students_common, 
                                 dfMath.shape[0]+dfPort.shape[0]-students_common))
```

```python
dfBoth = pd.concat([dfMath, dfPort]).drop_duplicates(subset=common_cols)
dfBoth.shape
```

By using the above approach, we were able to join both tables and remove duplicate records.

The difference between alcohol consumption on workdays and weekends is calculated as the difference between Dalc (workdays alcohol consumption) and Walc (weekends alcohol consumption) attributes. The negative values indicate an increase in the consumption of alcohol.

```python
dfBoth['AlcDiff'] = dfBoth['Dalc']-dfBoth['Walc']
dfBoth['AlcDiff'].head()
```

```python
dalc_diff = dfBoth.groupby(by='AlcDiff')['Dalc'].count().reset_index().rename(columns={'Dalc':'Count'})
dalc_diff
```

Plotting bar graphs and pie charts to visualize this information.

```python
inc_col = 'lightcoral'
same_col = 'lightskyblue'
dec_col = 'yellowgreen'

def apply_colors(d):
    if d['AlcDiff'] < 0:
        return inc_col
    elif d['AlcDiff'] == 0:
        return same_col
    else:
        return dec_col

colors = dalc_diff.apply(apply_colors, axis=1)
```

```python
plt.figure(figsize=(7,7))
plt.bar(x=dalc_diff['AlcDiff'], height=dalc_diff['Count'], color=colors, 
        tick_label=dalc_diff['AlcDiff'])
plt.xlabel('Alcohol Difference')
plt.ylabel('Number of students');
```

```python
plt.figure(figsize=(6,6))
plt.pie(x=[dalc_diff[dalc_diff['AlcDiff']<0]['Count'].sum(),
           dalc_diff[dalc_diff['AlcDiff']==0]['Count'].sum(),
           dalc_diff[dalc_diff['AlcDiff']>0]['Count'].sum()],
       labels=['Increase','Same','Decrease'], colors=['tomato',(0.17, 0.69, 0.97),'green'],
       autopct='%1.1f%%', pctdistance=0.8, textprops={'color':'white', 'fontweight':'bold'});
cir = plt.Circle((0,0), 0.6, color='black', fc='white', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(cir)
cir = plt.Circle((0,0), 1, ec='black', fill=False, linewidth=1)
fig.gca().add_artist(cir)
plt.legend(ncol=2, bbox_to_anchor=(0, 1), loc='lower left', fontsize='medium')
plt.axis('equal');
```

## Result

As evident from the pie chart, the proportion of students who increased their alcohol consumption is a little over 50% and only a small 1.8% of students decreased their intake.

## Question 3: Does parents’ education affects student’s grades?

My last question: is there a correlation between parents’ education and student’s grade. <br>For this analysis, we can simply concatenate the students’ records from both datasets because there are different grades for common students in different subjects. The total records in the relevant dataset become 1044.

```python
dfConcat = pd.concat([dfMath, dfPort])
print(dfConcat.shape)
dfConcat.head()
```

We categorize parents’ being highly educated as either father or mother possessing higher education (value 4 in Fedu and Medu fields) and use previous criteria of the good grade being greater than or equal to 3. Higher performance (HighPerf) is shown in red and lower performance (LowPerf) in blue.

```python
dfConcat['HighParentEdu'] = dfConcat.apply(lambda record: record['Fedu'] == 4 or record['Medu']==4, axis=1)
```

```python
highparedu = dfConcat[dfConcat['HighParentEdu']]
highparedu.shape
```

```python
highpareduPerformance = (highparedu.groupby('Performance').count()['G3']*100/
 highparedu.shape[0]).reset_index().rename(columns={'G3':'Proportion'})
highpareduPerformance
```

Visualizing the data:

```python
plt.figure(figsize=(10,1))
h = plt.barh(y=highpareduPerformance['Performance'][0], width=highpareduPerformance['Proportion'][0], 
             height=0.1, tick_label='', label='LowPerf', color='tomato',
            linewidth=1, edgecolor='black')
plt.barh(y=highpareduPerformance['Performance'][0], width=highpareduPerformance['Proportion'][1], 
         height=0.1, left=highpareduPerformance['Proportion'][0], label='HighPerf',
        linewidth=1, edgecolor='black')
plt.legend(ncol=2, bbox_to_anchor=(0, 1), loc='lower left', fontsize='medium')
plt.text(highpareduPerformance['Proportion'][0]/2,0, "{:.2f}".format(highpareduPerformance['Proportion'][0]), 
         color='white', fontsize='x-large', fontweight='heavy')
plt.text(highpareduPerformance['Proportion'][0]+(highpareduPerformance['Proportion'][1]/3),0, 
         "{:.2f}".format(highpareduPerformance['Proportion'][1]), 
         color='white', fontsize='x-large', fontweight='heavy')
plt.title("High Parents' Education")
plt.gca().set_axis_off()
```

```python
lowparedu = dfConcat[~dfConcat['HighParentEdu']]
lowparedu.shape
```

```python
lowpareduPerformance = (lowparedu.groupby('Performance').count()['G3']*100/
 lowparedu.shape[0]).reset_index().rename(columns={'G3':'Proportion'})
lowpareduPerformance
```

```python
plt.figure(figsize=(10,1))
h = plt.barh(y=lowpareduPerformance['Performance'][0], width=lowpareduPerformance['Proportion'][0], 
             height=0.1, tick_label='', label='LowPerf', color='tomato',
             linewidth=1, edgecolor='black')
plt.barh(y=lowpareduPerformance['Performance'][0], width=lowpareduPerformance['Proportion'][1], 
         height=0.1, left=lowpareduPerformance['Proportion'][0], label='HighPerf',
        linewidth=1, edgecolor='black')
plt.legend(ncol=2, bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')
plt.text(lowpareduPerformance['Proportion'][0]/2,0, "{:.2f}".format(lowpareduPerformance['Proportion'][0]), 
         color='white', fontsize='x-large', fontweight='heavy')
plt.text(lowpareduPerformance['Proportion'][0]+(lowpareduPerformance['Proportion'][1]/3),0, 
         "{:.2f}".format(lowpareduPerformance['Proportion'][1]), 
         color='white', fontsize='x-large', fontweight='heavy')
plt.title("Low Parents' Education")
plt.gca().set_axis_off()
plt.gca().set_xticks([0,50,100])
plt.xticks([0,50,100]);
```

## Result

The above bar graph illustrates that parents' education seems to have an impact on child's performance. Majority of students whose parents had higher education, performed better. And majority of students whose parents did not have higher education had low performances.
